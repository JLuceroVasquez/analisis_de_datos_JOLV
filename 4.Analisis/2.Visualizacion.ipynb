```python
#Importamos la libreria matplotlib para poder visualizar nuestros gráficos
import matplotlib.pyplot as plt
```

```python
#Gráficos de línea (relación entre 2 variables continuas, para representar tendencias)
X = [1,2,3,4,5]
Y = [2,4,7,6,8 ]
#Cuando solo añadimos 2 series, entonces plt asume que es un gráfico de línea
plt.plot(X,Y)
#Añadimos etiquetas a los ejes
plt.xlabel('edad')
plt.ylabel('altura')
#Añadimos el título del gráfico
plt.title('Edad vs altura')
plt.show
```

```python
#Gráfico de dispersión
plt.scatter(X,Y)
plt.title('Gráfico de dispersion')
plt.show()
```

```python
#Histogramas (visualizamos la frecuencia de los datos en diferentes intervalos, para detectar valores atípicos o patrones de comportamiento)
#Importamos numpy, se emplea en ciencia de datos. Y lo usaremos en este caso para generar datos
import numpy as np
datos = np.random.randn(1000)
#Ploteamos el histograma (bins= #rangos, color= color de las barras, )
plt.hist(datos, bins=30, color='skyblue', edgecolor='black')
plt.show()
```

```python
#Importamos la librería seaborn para emplear sus funciones
import seaborn as sn

#Confirgurar estilo en seaborn, nuestro fondo sera blanco
sn.set_theme(style='whitegrid')

#Configuramos nuestro histograma
sn.histplot(datos, color='skyblue')
plt.show()
```
