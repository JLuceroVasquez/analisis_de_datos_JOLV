```python
#Importamos la libreria matplotlib para poder visualizar nuestros gr√°ficos
import matplotlib.pyplot as plt
```

```python
#Gr√°ficos de l√≠nea (relaci√≥n entre 2 variables continuas, para representar tendencias)
X = [1,2,3,4,5]
Y = [2,4,7,6,8 ]
#Cuando solo a√±adimos 2 series, entonces plt asume que es un gr√°fico de l√≠nea
plt.plot(X,Y)
#A√±adimos etiquetas a los ejes
plt.xlabel('edad')
plt.ylabel('altura')
#A√±adimos el t√≠tulo del gr√°fico
plt.title('Edad vs altura')
plt.show
```

```python
#Gr√°fico de dispersi√≥n
plt.scatter(X,Y)
plt.title('Gr√°fico de dispersion')
plt.show()
```

```python
#Histogramas (visualizamos la frecuencia de los datos en diferentes intervalos, para detectar valores at√≠picos o patrones de comportamiento)
#Importamos numpy, se emplea en ciencia de datos. Y lo usaremos en este caso para generar datos
import numpy as np
datos = np.random.randn(1000)
#Ploteamos el histograma (bins= #rangos, color= color de las barras, )
plt.hist(datos, bins=30, color='skyblue', edgecolor='black')
plt.show()
```
