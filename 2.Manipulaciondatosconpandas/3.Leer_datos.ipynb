```python
import pandas as pd
import sqlalchemy as db
```

```python
#Vinculo al motor de BD para escribir SQL en Phyton
engine = db.create_engine("mysql://root:root@127.0.0.1:3310/retail_db")
#Obtiene la conexión "mysql://user:password@ip_servidor:puerto/database""
#Puedes conocer el usuario y contraseña revisa el dockerfile y el docker-compose.
#Puedes conocer el ip y puerto haciendo clic en database> icono del documento en la BD reatil_db (open query)
#En tu organización, la oficina de Tecnología te darán la ip_servidor y el puerto
conn = engine.connect()
```

```python
df_categories = pd.read_sql_query('select * from categories', con=conn)
df_categories.head()
```
