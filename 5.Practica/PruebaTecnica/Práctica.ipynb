```python
import pandas as pd
#Importamos el archivo csv, decodificamos empleanso ISO, indicamos que el separador es ';', y pedimos que omita la primera línea como registro (así lo toma como cabecera)
df = pd.read_csv('Resultado.csv', encoding='ISO-8859-1', sep=';', skiprows=1)
df.head(100)
#El problema es que no está separando los datos de fb e ig
```

```python
#Filtramos los datos por el índice autogenerado para separar ig de fb. Donde fb es hasta 28.
#El operador & es para operaciones bit a bit, y para aplicarlo correctamente a condiciones en un DataFrame, debes rodear cada condición con paréntesis.
#Ambas condiciones deben estar entre paréntesis completos.
df_filtrafo_facebook = df.loc[(df.index >= 0) & (df.index < 28)]
df_filtrafo_facebook
```

```python
#Filtramos los datos por el índice autogenerado para separar ig de fb. Donde Ig es apartir del dato 31.
df_filtrafo_instagram = df.loc[df.index > 30]
df_filtrafo_instagram = df_filtrafo_instagram.rename(columns={"Alcance de la página de Facebook": "alcance de la página de Instagram"})
df_filtrafo_instagram
```
