```python
#Read_csv trata de leer un archivo en formato UTF8. Si el archivo no fue guardado así, entonces arroja un error 'UnicodeDecodeError'.
sales_df = pd.read_csv("./data/sales_data_sample.csv", encoding='ISO-8859-1')
#Mostramos la cabecera del dataframesales
sales_df.head()
```

```python
#¿Cómo vemos los nulos en un df?
#Definimos una variable que arrje todos los registros nulos por columna
total_nulos = sales_df.isnull().sum().sort_values(ascending=False)
total_nulos
```

```python
#Definimos una variable que arrje el porcentaje de registros nulos por columna
var1 = sales_df.isnull().sum()
#Agregamos 'isnull' para que también cuente los registros nulos. Si no se especifica, count solo cuenta registros que tiene datos.
var2 = sales_df.isnull().count()
pct_nulos=var1/var2
pct_nulos
```
