```python
import pandas as pd
```

```python
#Dataframe a partir de lista
#Creamos la lista 'data'
data = [['Juan',25,'Dirección1'],['Maria',30,'Dirección2'],['Manuel',35,'Dirección3']]
#Creamos un dataframe df usando la función Dataframe de la librería de pandas
df = pd.DataFrame(data, columns=['Nombre','Edad','Dirección'])
df
#Al cargar los datos en el dataframe, la cantidad de columnas debe ser mayor o igual a la cantidad de valores ingresados dentro de cada lista embebida.
```

```python
#Dataframe a partir de Diccionario
#Creamos el diccionario 'data'. Cada elemento está compuesto por una clave y valor
data = {
    'Nombre':['Juan','Maria','Manuel' ],
    'Edad':[25,30,35],
}
#No es necesario especificar las columnas, pandas supone que las claves lo son.
df = pd.DataFrame(data)
df
#Las API comparten datos en formato JSON. A continuación, se realiza una transformación previa a convertir los datos en un dataframe.
```

```python
#Creamos un nuevo df a partir del diccioanrio dataletras
dataletras = {
    'Grupo':['A','B','A','A','B'],
    'Valor':[10,15,35,67,38]

}
df = pd.DataFrame(dataletras)
df
```

```python
promedio = df.groupby('Grupo').mean()
promedio
```

```python
promedio = df.groupby('Grupo').sum()
promedio
```

```python
df['Valor doble']=df['Valor'].apply(lambda x: x*2)
df
```

```python
df['Suma']=df.apply(lambda row: row['Valor']+ row['Valor doble'], axis=1)
df
```
